```python
import pandas as pd
import numpy as np

#files to read
data = "Sheets/world_happiness.csv"

#read yearly data files
wh = pd.read_csv(data)
whry = wh.loc[:,['Country or region',
                 'Score_15',
                 'Score_16',
                 'Score_17',
                 'Score_18',
                 'Score_19',
                 'Overall rank_15',
                 'Overall rank_16',
                 'Overall rank_17',
                 'Overall rank_18',
                 'Overall rank_19']]
whry.head()
#whry.to_csv("Sheets/whry.csv", index=False, header=True)
```

```python
whryg = whry.set_index("Country or region")
whryg.head()
```

```python
bemdf = pd.DataFrame(whryg.loc[['Argentina',
                 'Brazil',
                 'China',
                 'India',
                 'Indonesia',
                 'Mexico',
                 'Poland',
                 'South Africa',
                 'South Korea',
                 'Turkey'],:])
#whryg8.to_csv("Sheets/whryg8.csv", index=False, header=True)
#France, Germany, Italy, the United Kingdom, Japan, the United States, Canada, and Russia
# Argentina, Brazil, China, India, Indonesia, Mexico, Poland, South Africa, South Korea and Turkey
```

```python
bemdf['c_s'] = bemdf['Score_19'] - bemdf['Score_15']
bemdf['c_r'] = bemdf['Overall rank_19'] - bemdf['Overall rank_15']
bemdf['%c_s'] = (((bemdf['Score_19'] - bemdf['Score_15']) / bemdf['Score_15']) *100)
bemdf = bemdf.reset_index()
bemdf
```

```python
whry['c_s'] = whry['Score_19'] - whry['Score_15']
whry['c_r'] = whry['Overall rank_19'] - whry['Overall rank_15']
whry['%c_s'] = (((whry['Score_19'] - whry['Score_15']) / whry['Score_15']) *100)
whryna = whry.dropna()
```

```python
whryna
```

```python
whry['c_r'].max()
```

```python
whry_cr = whryna[['Country or region', 
                 'c_s',
                 'c_r',
                 '%c_s']]
whry_cr = whry_cr.sort_values(by=['c_r'])
whry_cr
```

```python
whry_cs = whryna[['Country or region', 
                 'c_s',
                 'c_r',
                 '%c_s']]
whry_cs = whry_cs.sort_values(by=['c_s'], ascending=False)
whry_cs
```
