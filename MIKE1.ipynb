```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import warnings
warnings.filterwarnings('ignore')
import plotly as py
import plotly.express as px

from plotly.offline import init_notebook_mode, iplot # plotly offline mode
init_notebook_mode(connected=True) 

# File to Load (Remember to Change These)
d2015 = "2015.csv"
d2016 = "2016.csv"
d2017 = "2017.csv"
d2018 = "2018.csv"
d2019 = "2019.csv"

# Read yearly data files
data_15 = pd.read_csv(d2015)
data_16 = pd.read_csv(d2016)
data_17 = pd.read_csv(d2017)
data_18 = pd.read_csv(d2018)
data_19 = pd.read_csv(d2019)
```

```python
# year 2015 extract columns and rearrange columns

df15c = data_15.loc[:,['Happiness Rank', 'Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Generosity',
       'Trust (Government Corruption)']]

#rename column
df15new= df15c.rename(columns={"Happiness Rank":"Overall rank",
                               "Country":"Country or region",
                               "Happiness Score":"Score",
                               "Economy (GDP per Capita)":"GDP per capita",
                               "Family":"Social support",
                               "Health (Life Expectancy)":"Healthy life expectancy",
                               "Freedom":"Freedom to make life choices",
                               "Generosity":"Generosity", 
                               "Trust (Government Corruption)": "Perceptions of corruption"})


df15new['Country or region'].replace(['Hong Kong S.A.R., China','Taiwan Province of China','Trinidad and Tobago','North Cyprus','Somaliland region'],['Hong Kong','Taiwan','Trinidad & Tobago','Northern Cyprus','Somaliland Region'], inplace = True)
df15new
```

```python
#year 2016 extract columns and rearrange columns

df16c = data_16.loc[:,['Happiness Rank', 'Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Generosity',
       'Trust (Government Corruption)']]

#rename column
df16new = df16c.rename(columns={"Happiness Rank":"Overall rank", "Country":"Country or region","Happiness Score":"Score","Economy (GDP per Capita)":"GDP per capita", "Family":"Social support", "Health (Life Expectancy)":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Generosity":"Generosity", "Trust (Government Corruption)": "Perceptions of corruption"})

#replacing name change for HK and Taiwan
df16new['Country or region'].replace(['Hong Kong S.A.R., China','Taiwan Province of China','Trinidad and Tobago','North Cyprus','Somaliland region'],['Hong Kong','Taiwan','Trinidad & Tobago','Northern Cyprus','Somaliland Region'], inplace = True)
df16new
```

```python
data_17.columns
```

```python
#year 2017 extract columns and rearrange columns

df17c = data_17.loc[:,['Happiness.Rank', 'Country', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Family',
       'Health..Life.Expectancy.', 'Freedom', 'Generosity',
       'Trust..Government.Corruption.']]

#rename column
df17new = df17c.rename(columns={"Happiness.Rank":"Overall rank", "Country":"Country or region","Happiness.Score":"Score","Economy..GDP.per.Capita.":"GDP per capita", "Family":"Social support", "Health..Life.Expectancy.":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Generosity":"Generosity", "Trust..Government.Corruption.": "Perceptions of corruption"})

#replacing name change for HK and Taiwan

df17new['Country or region'].replace(['Hong Kong S.A.R., China','Taiwan Province of China','Trinidad and Tobago','North Cyprus','Somaliland region'],['Hong Kong','Taiwan','Trinidad & Tobago','Northern Cyprus','Somaliland Region'], inplace = True)
df17new
```

```python
df1516 = pd.merge(df15new, df16new, how="outer", on = "Country or region", suffixes=(" 15", " 16"))
df1516
```

```python
df1718 = pd.merge(df17new, data_18, how="outer", on = "Country or region", suffixes=(" 17", " 18"))
df1718
```

```python
df17_19 = pd.merge(df1718, data_19, how="outer", on = "Country or region", suffixes=("", " 19"))
df17_19
```

```python
df_all= pd.merge(df1516, df17_19, how="outer", on = "Country or region", suffixes=("", ""))


df_all = df_all.rename(columns={"Overall rank":"Overall rank 19","Score":"Score 19","GDP per capita":"GDP per capita 19", "Social support":"Social support 19", "Healthy life expectancy":"Healthy life expectancy 19", "Freedom to make life choices":"Freedom to make life choices 19", "Generosity":"Generosity 19", "Perceptions of corruption": "Perceptions of corruption 19"})

df_all
```

```python
df_all.dropna(how='any',inplace=True)
df_all
```

```python
df_all.to_csv("world_happinesstam.csv", index=False, header=True)
```

```python
df_all['average_happiness_score_all'] = df_all[['Score 15', 'Score 16','Score 17', 'Score 18','Score 19']].mean(axis=1)
df_all
```

```python
df_all['average_rank_all'] = df_all[['Overall rank 15',
                                     'Overall rank 16',
                                     'Overall rank 17', 
                                     'Overall rank 18',
                                     'Overall rank 19']].mean(axis=1)

df_all['average_GDP_all'] = df_all[['GDP per capita 15',
                                     'GDP per capita 16',
                                     'GDP per capita 17', 
                                     'GDP per capita 18',
                                     'GDP per capita 19']].mean(axis=1)

df_all['average_social_all'] = df_all[['Social support 15',
                                     'Social support 16',
                                     'Social support 17', 
                                     'Social support 18',
                                     'Social support 19']].mean(axis=1)

df_all['average_health_all'] = df_all[['Healthy life expectancy 15',
                                     'Healthy life expectancy 16',
                                     'Healthy life expectancy 17', 
                                     'Healthy life expectancy 18',
                                     'Healthy life expectancy 19']].mean(axis=1)

df_all['average_freedom_all'] = df_all[['Freedom to make life choices 15',
                                     'Freedom to make life choices 16',
                                     'Freedom to make life choices 17', 
                                     'Freedom to make life choices 18',
                                     'Freedom to make life choices 19']].mean(axis=1)

df_all['average_generosity_all'] = df_all[['Generosity 15',
                                     'Generosity 16',
                                     'Generosity 17', 
                                     'Generosity 18',
                                     'Generosity 19']].mean(axis=1)

df_all['average_corruption_all'] = df_all[['Perceptions of corruption 15',
                                     'Perceptions of corruption 16',
                                     'Perceptions of corruption 17', 
                                     'Perceptions of corruption 18',
                                     'Perceptions of corruption 19']].mean(axis=1)
df_all
```

```python
df_all.to_csv("world_happinesstam.csv", index=False, header=True)
```

```python
top_5_country = df_allavg.sort_values('average_happiness_score_all', ascending= False).head(5)['Country or region']
generosity_rank = [np.where(df_allavg.sort_values('average_generosity_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
gdp_rank = [np.where(df_allavg.sort_values('average_GDP_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
Social_Support_rank = [np.where(df_allavg.sort_values('average_social_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
Healthy_life_exp_rank = [np.where(df_allavg.sort_values('average_health_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
Freedom_choice__rank = [np.where(df_allavg.sort_values('average_freedom_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
Perce_corruption_rank = [np.where(df_allavg.sort_values('average_corruption_all', ascending= False).reset_index()['Country or region']==i)[0][0] +1 for i in top_5_country]
```

```python
feature_rank_top_5_country = pd.DataFrame({
    'country':top_5_country,
    'Generosity_rank':generosity_rank,
    'GDP_rank':gdp_rank,
    'Social_Support_rank':Social_Support_rank,
    'Healthy_life_expectancy_rank':Healthy_life_exp_rank,
    'Freedom_make_choices_rank':Freedom_choice__rank,
    'Perceptions_of_corruption_rank':Perce_corruption_rank})
feature_rank_top_5_country.to_csv("top5rankavg.csv", index=False, header=True)
```
