```python
# Dependencies and Setup
import pandas as pd

import numpy as np

# File to Load (Remember to Change These)
d2015 = "Sheets/2015.csv"
d2016 = "Sheets/2016.csv"
d2017 = "Sheets/2017.csv"
d2018 = "Sheets/2018.csv"
d2019 = "Sheets/2019.csv"

# Read yearly data files
data_15 = pd.read_csv(d2015)
data_16 = pd.read_csv(d2016)
data_17 = pd.read_csv(d2017)
data_18 = pd.read_csv(d2018)
data_19 = pd.read_csv(d2019)

#year 2015 extract columns and rearrange columns
df15c = data_15.loc[:,['Country',
                       'Happiness Rank', 
                       'Happiness Score', 
                       'Economy (GDP per Capita)',
                       'Family',
                       'Health (Life Expectancy)', 
                       'Freedom', 
                       'Generosity',
                       'Trust (Government Corruption)']]

#rename columns
df15new= df15c.rename(columns={"Happiness Rank":"Rank", 
                               "Country":"Country",
                               "Happiness Score":"Score",
                               "Economy (GDP per Capita)":"GDP", 
                               "Family":"Family", 
                               "Health (Life Expectancy)":"Health",
                               "Freedom":"Freedom",
                               "Generosity":"Generosity", 
                               "Trust (Government Corruption)":"Trust"})

#replace country names 
df15new['Country'].replace(['Hong Kong S.A.R., China',
                                      'Taiwan Province of China',
                                      'Trinidad and Tobago',
                                      'North Cyprus','Somaliland region'],
                                     ['Hong Kong',
                                      'Taiwan',
                                      'Trinidad & Tobago',
                                      'Northern Cyprus',
                                      'Somaliland Region'], inplace = True)

#year 2016 extract columns and rearrange columns

df16c = data_16.loc[:,['Country',
                       'Happiness Rank',
                       'Happiness Score',
                       'Economy (GDP per Capita)', 
                       'Family',
                       'Health (Life Expectancy)',
                       'Freedom', 'Generosity',
                       'Trust (Government Corruption)']]

#rename column
df16new = df16c.rename(columns={"Happiness Rank":"Rank",
                                "Country":"Country",
                                "Happiness Score":"Score",
                                "Economy (GDP per Capita)":"GDP",
                                "Family":"Family",
                                "Health (Life Expectancy)":"Health",
                                "Freedom":"Freedom",
                                "Generosity":"Generosity",
                                "Trust (Government Corruption)": "Trust"})

#replacing name change for HK and Taiwan
df16new['Country'].replace(['Hong Kong S.A.R., China',
                                      'Taiwan Province of China',
                                      'Trinidad and Tobago',
                                      'North Cyprus',
                                      'Somaliland region'],
                                     ['Hong Kong',
                                      'Taiwan',
                                      'Trinidad & Tobago',
                                      'Northern Cyprus',
                                      'Somalia'], inplace = True)


#year 2017 extract columns and rearrange columns

df17c = data_17.loc[:,['Country',
                       'Happiness.Rank',
                       'Happiness.Score',
                       'Economy..GDP.per.Capita.',
                       'Family',
                       'Health..Life.Expectancy.',
                       'Freedom',
                       'Generosity',
                       'Trust..Government.Corruption.']]

#rename column
df17new = df17c.rename(columns={"Country":"Country",
                                "Happiness.Rank":"Rank",
                                "Happiness.Score":"Score",
                                "Economy..GDP.per.Capita.":"GDP", 
                                "Family":"Family", 
                                "Health..Life.Expectancy.":"Health",
                                "Freedom":"Freedom",
                                "Generosity":"Generosity", 
                                "Trust..Government.Corruption.": "Trust"})

df18c = data_18.loc[:,['Country or region',
                       'Overall rank', 
                       'Score',
                       'GDP per capita',
                       'Social support',
                       'Healthy life expectancy',
                       'Freedom to make life choices',
                       'Generosity',
                       'Perceptions of corruption']]

df18new = df18c.rename(columns={"Overall rank":"Rank",
                                "Country or region":"Country",
                                "Score":"Score",
                                "GDP per capita":"GDP", 
                                "Social support":"Family", 
                                "Healthy life expectancy":"Health",
                                "Freedom to make life choices":"Freedom",
                                "Generosity":"Generosity", 
                                "Perceptions of corruption": "Trust"})

df19c = data_19.loc[:,['Country or region',
                       'Overall rank', 
                       'Score',
                       'GDP per capita',
                       'Social support',
                       'Healthy life expectancy',
                       'Freedom to make life choices',
                       'Generosity',
                       'Perceptions of corruption']]

df19new = df19c.rename(columns={"Overall rank":"Rank",
                                "Country or region":"Country",
                                "Score":"Score",
                                "GDP per capita":"GDP", 
                                "Social support":"Family", 
                                "Healthy life expectancy":"Health",
                                "Freedom to make life choices":"Freedom",
                                "Generosity":"Generosity", 
                                "Perceptions of corruption": "Trust"})

# replacing name change for HK and Taiwan

df17new['Country'].replace(['Hong Kong S.A.R., China',
                                      'Taiwan Province of China',
                                      'Trinidad and Tobago',
                                      'North Cyprus',
                                      'Somaliland region'],
                                     ['Hong Kong',
                                      'Taiwan',
                                      'Trinidad & Tobago',
                                      'Northern Cyprus',
                                      'Somaliland Region'], inplace = True)
```

```python
df1516 = pd.merge(df15new, df16new, how="outer", on = "Country", suffixes=("_15", "_16"))

df1718 = pd.merge(df17new, df18new, how="outer", on = "Country", suffixes=("_17", "_18"))

df17_19 = pd.merge(df1718, df19new, how="outer", on = "Country")

df_all= pd.merge(df1516, df17_19, how="outer", on = "Country")
df_all
```

```python
df_alln = df_all.rename(columns={"Rank":"Rank_19",
                                "Score":"Score_19",
                                "GDP":"GDP_19", 
                                "Family":"Family_19", 
                                "Health":"Health_19",
                                "Freedom":"Freedom_19",
                                "Generosity":"Generosity_19", 
                                "Trust": "Trust_19"})
```
