```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import warnings
warnings.filterwarnings('ignore')
import plotly as py
import plotly.graph_objs as go
import pandas as pd

from plotly.offline import init_notebook_mode, iplot # plotly offline mode
init_notebook_mode(connected=True) 

file = 'sheets/happiness_avg.csv'
avgdf = pd.read_csv(file)
avgdf
```

```python
#compute the correlation matrix among the variables and plotted as heat map.

avg_matrix = avgdf[['Avg. Happiness',
                    'Avg. GDP',
                    'Avg. Family',
                    'Avg. Health',
                    'Avg. Freedom',
                    'Avg. Generosity',
                    'Avg. Corruption']]
cor = avg_matrix.corr() 
plt.savefig('Matrix Happiness Score.png')
sns.heatmap(cor, annot = True, fmt= '.2f')
plt.rcParams['figure.figsize'] = [20.0, 7.0]
plt.title('Matrix Happiness Correlation')
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.savefig('Matrix Happiness Score2.png')
plt.show() # ta-da!
#sns.heatmap(cor, square = True) 
```
